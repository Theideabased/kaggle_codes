#                                   **Titanic survival prediction**#
To predict the people that survive in titanic ship wreck we will have to view some data set given to us and use this to predict. we will use the train dataset to build the model and the test dataset to make sample prediction.
the list of the dataset are :
1. train dataset: this will be use to build our model 
2. test dataset: this will be use to check our model
3. gender submission dataset: this is an asumption dataset that shows that all female in the ship survive. 

```python
# let's load our train dataset
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
#Let's load our test dataset
test_data = pd.read_csv("/kaggle/input/titanic/test.csv") 
test_data.head()
```

Now that we have load our train data and our test data let's check how true is he gender submission data and let see if gender is a good determinate of survival 

```python
#let's check the rate of women that survive
women = train_data.loc[train_data.Sex == 'female'] ["Survived"]
rate_women = sum(women)/len(women)
print("% of women that survive = ", rate_women)
```

```python
# Let's check the rate of men that survive
men = train_data.loc[train_data.Sex == "male"] ["Survived"]
rate_men = sum(men)/len(men)
print("% of men that survive = ", rate_men)
```

As we can see the gender submisssion prediction is not accurate but it is a strong determinate for surviving however we can not analyse all the factors one after the other so we will build a machine learning model to analyse our data.
And make the prediction accordingly 

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex","SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth= None, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
